```python
# Import Fashion dataset
df1 = pd.read_csv('/content/fashion dataset.csv')
df1
```

```python
# Merging both datasets
df1 = pd.read_csv('/content/fashion dataset.csv')
df2 = pd.read_csv('/content/fashion brand details.csv')

df_merged = pd.merge (df1,df2, left_on='brand', right_on='brand_name', how='left')
df_merged

```

```python
# using the 'pandas.DataFrame.to_csv' method used earlier, i saved the merged dataset to a CSV file:

df_merged.to_csv('aaliyahs_merged_fashion_dataset.csv', index=False)
df_merged
```

```python
# To confirm that the current position of our columns, we can print a list of all the
# columns in the dataframe using the df_merged.columns attribute:

print(df_merged.columns)
```

```python
# using the 'pandas.DataFrame.pop' and 'pandas.DataFrame.insert' method to remove and reposition
# the 'brand_id' column on the dataset

brand_id = df_merged.pop('brand_id')

df_merged.insert(1, 'brand_id', brand_id)
```

```python
# To confirm that the column has been moved to our desired position, we can print a list of all the
# columns in the dataframe using the 'df_merged.columns' attribute:

print(df_merged.columns)
```

```python
# using the 'pandas.DataFrame.pop' and 'pandas.DataFrame.insert' method to remove and reposition
# the 'brand_name' column on the dataset

brand_name = df_merged.pop('brand_name')

df_merged.insert(2, 'brand_name', brand_name)
```

```python
# using the 'pandas.DataFrame.pop' and 'pandas.DataFrame.insert' method to remove and reposition
# the 'brand' column on the dataset

brand = df_merged.pop('brand')

df_merged.insert(3, 'brand', brand)
```

```python
# To confirm that the column has been moved to our desired position, we can print a list of all the
# columns in the dataframe using the 'df_merged.columns' attribute:

print(df_merged.columns)
```

```python
# print the merged dataset to view our repositioned columns

df_merged
```

```python

# To clean our dataset, we will delete irrelevant columns, we use the 'pandas.DataFrame.drop' method
#  The columns to be dropped are 'p_id', 'brand_name', 'description', 'p_attributes'

df_merged = df_merged.drop(columns=['p_id', 'brand_name', 'description', 'p_attributes'])
df_merged
```

```python
# To proceed with cleaninig our dataset, we will need to replace all Null or NaN values with 0, to do this we will be using the
# 'pandas.DataFrame.fillna' method

# Replace NaN values with 0
df_merged = df_merged.fillna(0)
df_merged

```

```python
# After removing the null and NaN values, we will proceed to removing duplicate rows

df_merged = df_merged.drop_duplicates()
df_merged
```

```python
# using the 'pandas.DataFrame.to_csv' method to save our cleaned dataset to a CSV file:

df_merged.to_csv('Cleaned_aaliyahs_merged_fashion_dataset.csv', index=False)
df_merged
```
